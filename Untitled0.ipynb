```python
# Data Handling and Cleaning
# Dropping rows with missing essential information for Age, Height, Weight, or Medal
cleaned_data = df.dropna(subset=['Age', 'Height', 'Weight', 'Medal'])
cleaned_data
```

```python
# Univariate Analysis - Plot distributions for Age, Height, Weight, and Medal counts
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=axs[0, 0]).set_title('Age Distribution')
sns.histplot(df['Height'].dropna(), bins=30, kde=True, ax=axs[0, 1]).set_title('Height Distribution')
sns.histplot(df['Weight'].dropna(), bins=30, kde=True, ax=axs[1, 0]).set_title('Weight Distribution')
sns.countplot(data=df, x='Medal', order=['Gold', 'Silver', 'Bronze'], ax=axs[1, 1]).set_title('Medal Counts')

plt.tight_layout()

```

```python
# Bivariate Analysis - Analyzing correlation between Age, Height, Weight with Medal achievement
bivariate_age_medal = sns.boxplot(data=cleaned_data, x='Medal', y='Age').set_title('Age vs Medal')
plt.figure()
bivariate_height_medal = sns.boxplot(data=cleaned_data, x='Medal', y='Height').set_title('Height vs Medal')
plt.figure()
bivariate_weight_medal = sns.boxplot(data=cleaned_data, x='Medal', y='Weight').set_title('Weight vs Medal')
```

```python
# Multivariate Analysis - Age, Height, Weight, and Sex distribution by Medal type
sns.pairplot(cleaned_data[['Age', 'Height', 'Weight', 'Medal', 'Sex']], hue='Medal', diag_kind='kde')
```
